```python
washing_machine = ['5A-YB7Z-ZFXP','Washing_Machine_Filter','201']
washing_machine_cartridge = ['RW-0JGB-G59A','Washing_Machine_Cartridge']

shower_filter_10stage = ['New-103 A' , '104_Single', 'New-103_1', 'New-103', 'New-104']
shower_filter_10_stage_cartridge = ['103 C']


black_shower_filter = ['1Z-ZD5D-W1P5','ZF-FXQE-MFJY','New_901','901']
black_shower_filter_cartridge =['901_Cartidge']

zeus = ['PF3SPRY-01']
zeus_cartridge =['3SPRY-C']

shower_head = ['OK-H0DX-9N0N','109']
shower_head_cartridge = ['109_Cartridge']

alkaline_filter =['Alkaline-01']
alkaline_cartridge = ['105']


tap_filter = ['Tabs_p_06','Tabs_p_03','PFKTF-01']
```

```python
df_wm = pd.DataFrame()
df_wm_cartridge = pd.DataFrame()

df_10stage =pd.DataFrame()
df_10stage_cartridge =pd.DataFrame()

df_black =pd.DataFrame()
df_black_cartridge =pd.DataFrame()

df_zeus= pd.DataFrame()
df_zeus_cartridge= pd.DataFrame()

df_shower_head= pd.DataFrame()
df_shower_head_cartridge= pd.DataFrame()

df_alkaline= pd.DataFrame()
df_alkaline_cartridge= pd.DataFrame()

df_tap_filter = pd.DataFrame()
```

```python
# for i in Washing_machine:
for row in range(df.shape[0]):
  if df.iloc[row]['SKU'] in washing_machine:
    df_wm = df_wm.append(df.iloc[row])

  if df.iloc[row]['SKU'] in washing_machine_cartridge:
    df_wm_cartridge = df_wm_cartridge.append(df.iloc[row])  

  if df.iloc[row]['SKU'] in shower_filter_10stage:
    df_10stage = df_10stage.append(df.iloc[row])

  if df.iloc[row]['SKU'] in shower_filter_10_stage_cartridge:
    df_10stage_cartridge = df_10stage_cartridge.append(df.iloc[row])    


  if df.iloc[row]['SKU'] in black_shower_filter:
    df_black = df_black.append(df.iloc[row])

  if df.iloc[row]['SKU'] in black_shower_filter_cartridge:
    df_black_cartridge = df_black_cartridge.append(df.iloc[row])


  if df.iloc[row]['SKU'] in zeus:
    df_zeus = df_zeus.append(df.iloc[row])    


  if df.iloc[row]['SKU'] in zeus_cartridge:
    df_zeus_cartridge = df_zeus_cartridge.append(df.iloc[row])

  if df.iloc[row]['SKU'] in shower_head:
    df_shower_head = df_shower_head.append(df.iloc[row])

  if df.iloc[row]['SKU'] in shower_head_cartridge:
    df_shower_head_cartridge = df_shower_head_cartridge.append(df.iloc[row])


  if df.iloc[row]['SKU'] in alkaline_filter:
    df_alkaline = df_alkaline.append(df.iloc[row])

  if df.iloc[row]['SKU'] in alkaline_cartridge:
    df_alkaline_cartridge = df_alkaline_cartridge.append(df.iloc[row])

  if df.iloc[row]['SKU'] in tap_filter:
    df_tap_filter = df_tap_filter.append(df.iloc[row])
```
