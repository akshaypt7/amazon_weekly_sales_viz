```python
df = pd.read_csv(io.BytesIO(uploaded_file['BusinessReport.csv']))


try:

  df.drop([ 'Sessions',
        'Session Percentage', 'Page views', 'Page Views Percentage',
        'Featured Offer (Buy Box) Percentage', 'Total Order Items',
        'Unit Session Percentage','(Parent) ASIN'],axis=1,inplace= True)
  
except:
  pass  
```

```python
# Nemora

kattil = ['Decorative-Light-01','HY-LDOK-X8MQ']

light_kattil = ['NMP03-WY']

koodu = ['NMP02-DW','Decorative-Light-02','ZJ-JU30-M8TR','NMP02-DW-WB']

light_koodu = ['NMP02-WY-WB','NMP02-WY','NMP02-WD-WB']

pyramid = ['NMP14-WD-WB','NMP14-WD']

triangle = ['NMP10-RW']

lap_top_stand = ['NMFT07-WD']

head_set_stand = ['NMFT04-WD']


nemora_products = ['kattil','light_kattil','koodu','light_koodu','pyramid','triangle','lap_top_stand','head_set_stand']
```

```python
df_kattil = pd.DataFrame()
df_light_kattil = pd.DataFrame()

df_koodu =pd.DataFrame()
df_light_koodu =pd.DataFrame()

df_pyramid =pd.DataFrame()

df_triangle= pd.DataFrame()

df_lap_top_stand= pd.DataFrame()

df_head_set_stand= pd.DataFrame()
```

```python
for row in range(df.shape[0]):
  if df.iloc[row]['SKU'] in kattil:
    df_kattil = df_kattil.append(df.iloc[row])

  if df.iloc[row]['SKU'] in light_kattil:
    df_light_kattil = df_light_kattil.append(df.iloc[row])

  if df.iloc[row]['SKU'] in koodu:
    df_koodu = df_koodu.append(df.iloc[row])

  if df.iloc[row]['SKU'] in light_koodu:
    df_light_koodu = df_light_koodu.append(df.iloc[row])    

  if df.iloc[row]['SKU'] in pyramid:
    df_pyramid = df_pyramid.append(df.iloc[row]) 


  if df.iloc[row]['SKU'] in triangle:
    df_triangle = df_triangle.append(df.iloc[row]) 


  if df.iloc[row]['SKU'] in lap_top_stand:
    df_lap_top_stand = df_lap_top_stand.append(df.iloc[row]) 

  if df.iloc[row]['SKU'] in head_set_stand:
    df_head_set_stand = df_head_set_stand.append(df.iloc[row])     
```
